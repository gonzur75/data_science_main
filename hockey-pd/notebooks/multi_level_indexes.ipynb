```python
team_splits.head()
```

```python
team_splits.drop(columns=['tmID'], inplace=True)
```

```python
team_splits.set_index(['year', 'name'], inplace=True)
```

```python
team_splits = team_splits.loc[:, 'SepW':'AprOL'].copy()
```

```python
team_splits.head()
```

```python
months = team_splits.columns.map(lambda x: x[:3])  #
```

```python
metrics = team_splits.columns.map(lambda x: x[3:])
```

```python
months
```

```python
metrics
```

```python
multi_index = pd.MultiIndex.from_arrays([months, metrics])
```

```python
multi_index
```

```python
team_splits.columns = multi_index
team_splits.head()
```

```python
team_splits = team_splits.stack(level=0, future_stack=True)
```

```python
team_splits = team_splits.reorder_levels([2, 0, 1])
```

```python
team_splits.head()
```

```python
team_splits.index.name
```

```python
team_splits
```

```python
team_splits = team_splits.swaplevel(0, 2)
```

```python
team_splits.head()
```

```python
team_splits.index.set_names('month', level=2, inplace=True)
team_splits.index.set_names('name', level=0, inplace=True)
```

```python
team_splits
```

```python
# sumujemy wszystkie wiersze dla każdego zespołu w każdym roku
team_splits.groupby(level=['name', 'year']).sum().head(10)
# team_splits.groupby(level=['name', 'year']).sum().sum(axis=1)
```
